```python
import json
import csv
import pandas as pd

# Opening JSON file and loading the data
# into the variable data
with open('The_Input_json_file.json') as json_file:
    data = json.load(json_file) 
```

```python
len(data)
```

```python
#informations that i needed to extract from json, creating list for them
No_name=[]
iD=[]
url=[]
topic=[]
title=[]
pubdate=[]
sentence=[]
outlet=[]
sentence_id=[]
sentence_type=[]
speaker_entity=[]
person_affiliation=[]
person_aff_type=[]
org_aff_type=[]
wdata=[]
notes=[]
```

```python
for i in range (len(data)):
    result_len= len(data[i]["annotations"][0]["result"])
    No_name.append(data[i]["data"]["Unnamed: 0"])
    iD.append(data[i]["data"]["id"])
    url.append(data[i]["data"]["url"])
    topic.append(data[i]["data"]["topic"])
    title.append(data[i]["data"]["title"])
    pubdate.append(data[i]["data"]["pubdate"])
    sentence.append(data[i]["data"]["sentence"])
    outlet.append(data[i]["data"]["outlet"])
    sentence_id.append(data[i]["data"]["sentence_id"])
    temp_list=[]
#     from_name == label
    for j in range(result_len): 
        if data[i]["annotations"][0]["result"][j]["from_name"]=="label":
            temp_list.append(data[i]["annotations"][0]["result"][j]["value"]["choices"][0])    
        else:
             continue
    if len(temp_list)<1:
        temp_list.append("N/A")
        sentence_type.append(temp_list)
    else:
        sentence_type.append(temp_list)
    temp_list=[]
#  from_name == speaker
    for k in range(result_len):
        if data[i]["annotations"][0]["result"][k]["from_name"]=="speaker":
            speaker_entity_dict={}
            key=data[i]["annotations"][0]["result"][k]["value"]["text"]
            value=data[i]["annotations"][0]["result"][k]["value"]["labels"][0]
            speaker_entity_dict[key]=value
            temp_list.append(speaker_entity_dict)
        else:
            continue
    if len(temp_list)==0:
        temp_list.append("N/A")
        speaker_entity.append(temp_list)
    else:
        speaker_entity.append(temp_list)
    temp_list=[]
#  from_name == person_affiliation
    for l in range(result_len):            
        if data[i]["annotations"][0]["result"][l]["from_name"]=="person_affiliation":
            temp_list.append(data[i]["annotations"][0]["result"][l]["value"]["choices"][0])
        else:
            continue 
    if len(temp_list)==0:
        temp_list.append("N/A")
        person_affiliation.append(temp_list)
    else:
        person_affiliation.append(temp_list)
    temp_list=[]
#  from_name == person_afftype
    for m in range(result_len):
        if data[i]["annotations"][0]["result"][m]["from_name"]=="person_afftype":
            temp_list.append(data[i]["annotations"][0]["result"][m]["value"]["choices"][0])
        else:
            continue
    if len(temp_list)==0:
        temp_list.append("N/A")
        person_aff_type.append(temp_list)
    else:
        person_aff_type.append(temp_list)
    temp_list=[]
#     from_name == orgtype
    for n in range(result_len):
        if data[i]["annotations"][0]["result"][n]["from_name"]=="orgtype":
            temp_list.append(data[i]["annotations"][0]["result"][n]["value"]["choices"][0])
        else:
            continue
    if len(temp_list)==0:
        temp_list.append("N/A")
        org_aff_type.append(temp_list)
    else:
        org_aff_type.append(temp_list)
    temp_list=[]
#     from_name == wdata           
    for o in range(result_len):
        if data[i]["annotations"][0]["result"][o]["from_name"]=="wdata":
            temp_list.append(data[i]["annotations"][0]["result"][o]["value"]["text"][0])
        else:
            continue
    if len(temp_list)==0:
        temp_list.append("N/A")
        wdata.append(temp_list)
    else:
        wdata.append(temp_list)
    temp_list=[]
  #     from_name == notes    
    for p in range(result_len):
        if data[i]["annotations"][0]["result"][p]["from_name"]=="notes":    
            temp_list.append(data[i]["annotations"][0]["result"][p]["value"]["text"][0])
        else:
            continue
    if len(temp_list)==0:
        temp_list.append("N/A")
        notes.append(temp_list)
    else:
        notes.append(temp_list)
    temp_list=[]

    
```

```python
# dictionary of lists  
dict = {'No_name': No_name, 'iD': iD, 'url': url, 'topic': topic, 'title': title, 'sentence': sentence, 'pubdate': pubdate, 'outlet': outlet, 'sentence_id': sentence_id, 'sentence_type': sentence_type, 'speaker_entity': speaker_entity,  'person_affiliation': person_affiliation, 'person_aff_type': person_aff_type, 'org_aff_type': org_aff_type, 'wdata': wdata, 'notes': notes}  
#       
df = pd.DataFrame(dict) 
```
